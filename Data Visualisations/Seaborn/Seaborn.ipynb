# üåü Seaborn

1. **üéõÔ∏è Layer of Abstraction**:
   - **Simpler to Use**: Seaborn simplifies complex plotting operations, making it easier to create aesthetically pleasing plots with less code. It abstracts much of the complexity involved in creating plots with Matplotlib.

2. **üé® Better Aesthetics**:
   - **Enhanced Visual Appeal**: Seaborn comes with a set of default themes and color palettes that make plots look more professional and visually appealing. It improves the look of plots with minimal effort from the user.

3. **üìä More Graph Types**:
   - **Wide Range of Plots**: Seaborn includes several built-in plot types that are not available in Matplotlib. This includes specialized plots for statistical analysis, such as violin plots, boxen plots, and pair plots.

## üó∫Ô∏è Seaborn Roadmap

Seaborn provides various functions that can be categorized into different types based on their functionality. Here‚Äôs an overview of these types and their main classifications:

## üîß Types of Functions

1. **üñºÔ∏è Figure Level Functions**:
   - Functions that operate on the level of an entire figure and manage the creation of subplots. These functions allow complex plots to be built more easily by setting up the structure and layout of the plot.

2. **üìè Axis Level Functions**:
   - Functions that draw onto a single matplotlib Axes. These functions are more granular and allow for fine-tuning of individual plot elements.

## üìÇ Main Classification

1. **üîó Relational Plot**:
   - **Examples**: `scatterplot`, `lineplot`
   - **Use**: To show relationships between two or more variables.

2. **üìà Distribution Plot**:
   - **Examples**: `histplot`, `kdeplot`, `ecdfplot`
   - **Use**: To show the distribution of a single variable.

3. **üìä Categorical Plot**:
   - **Examples**: `boxplot`, `violinplot`, `stripplot`, `swarmplot`
   - **Use**: To show the distribution of a variable across different categories.

4. **üìâ Regression Plot**:
   - **Examples**: `regplot`, `lmplot`
   - **Use**: To show the relationship between two variables along with a fitted regression model.

5. **üó∫Ô∏è Matrix Plot**:
   - **Examples**: `heatmap`, `clustermap`
   - **Use**: To show data in a matrix format, often used for correlation matrices or clustered heatmaps.

6. **üñºÔ∏è Multiplots**:
   - **Examples**: `FacetGrid`, `PairGrid`
   - **Use**: To create multiple plots that share a common structure and are used to visualize complex datasets.

For more detailed information about each function and its usage, you can refer to the [Seaborn API documentation](https://seaborn.pydata.org/api.html).

# Categorical Plots ‚ö™Ô∏è

## Strip plot üî≥
- A strip plot, also known as a dot plot or dot chart, is a type of data visualization that displays individual data points along a one-dimensional axis. 
- Each data point is represented by a dot, which is positioned along the axis according to its numerical value.

### Axes level function üî∂

```python
sns.stripplot(data=tips,x='day',y='total_bill')
```

### Using catplot figure level function üü§

```python
sns.catplot(data=tips, x='day',y='total_bill',kind='strip')
```

## Jitter üü°
- Adds a small amount of random jitter to the data points along the categorical axis ('day'), which helps in visualizing overlapping data points.

```python
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex') #With jitter
```

```python
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',hue='sex')  #Without jitter
```

## Swarmplot üî∂
- Swarm plots are similar to strip plots, but they adjust the position of points to avoid overlapping, making it easier to see individual data points


### Figure Function üî¥

```python
sns.catplot(data=tips, x='day',y='total_bill',kind='swarm')
```

```python
sns.catplot(data=tips, x='day',y='total_bill')
```

## Boxplot üü†

- A boxplot is a standardized way of displaying the distribution of data based on a five number summary (‚Äúminimum‚Äù, first quartile [Q1], median, third quartile [Q3] and ‚Äúmaximum‚Äù). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

### Single boxplot üî≥

```python
sns.boxplot(data=tips,y='total_bill')
```

## Violinplot ‚ñ™Ô∏è
- Violinplot = (Boxplot + KDEplot)
- **KDE Plot:** The shape of the violin plot shows the distribution density of the data.
- **Box Plot:** Inside the violin plot, the box plot shows the summary statistics (median, quartiles, and sometimes whiskers).

### Axes Level Fuction ‚òëÔ∏è

```python
sns.violinplot(data=tips,x='day',y='total_bill')
```

### Figure Level Fuction ‚òëÔ∏è

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True) #Hue
```

# Barplot ‚òëÔ∏è
A bar plot is a graphical representation of data where the height (or length) of bars is proportional to the values they represent. When combined with error bars, it provides a way to visualize the variability or uncertainty in the data.

- some issue with errorbar parameter

```python
import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill',ci=None)
```

# Point plot üü†

```python
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker',ci=None)
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars
